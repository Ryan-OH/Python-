# 데이터 불러오기
* **`pd.read_csv(csv파일 경로, parse_dates=[datetime형식으로 불러올 컬럼])`**

* 속성 정보
    * `datetime` - 시간대별 데이터 제공. timestamp(날짜와 시간 결합) 형태
    * `season`  - 1 = spring, 2 = summer, 3 = fall, 4 = winter
    * `holiday` - 공휴일(주말이 아닌 빨간날). 1-yes, 2-no
    * `workingday` - 주말, 공휴일을 제외한 날 1-yes, 2-no
    * `weather`
        * 1: Clear, Few clouds, Partly cloudy, Partly cloudy
        * 2: Mist + Cloudy, Mist + Broken clouds, Mist + Few clouds, Mist
        * 3: Light Snow, Light Rain + Thunderstorm + Scattered clouds, Light Rain + Scattered clouds
        * 4: Heavy Rain + Ice Pallets + Thunderstorm + Mist, Snow + Fog
    * `temp` - 섭씨온도
    * `atemp` - 섭씨체감온도
    * `humidity` - 상대습도
    * `windspeed` - 풍속
    * `casual` - 비회원 대여량
    * `registered` - 회원 대여량
    * `count` - 총 대여량


```python
# datetime 컬럼은 datetime 형식으로 읽어오도록 옵션 설정
df = pd.read_csv('https://raw.githubusercontent.com/JayoungKim-ai/ML_dataset/main/BikeSharingDemand.csv', parse_dates=['datetime'])
# df = pd.read_csv('data/bike_sharing_demand.csv', parse_dates=['datetime'])
display(df.head(3))
display(df.tail(3))
```

```python
# 데이터 정보
df.info()
```

# 파생컬럼 추가

```python
# 연, 월, 일, 시, 요일
df['year'] = df['datetime'].dt.year
df['month'] = df['datetime'].dt.month
df['day'] = df['datetime'].dt.day
df['hour'] = df['datetime'].dt.hour
df['dayofweek'] = df['datetime'].dt.dayofweek # 월요일:0, 일요일:6
df.head()
```

# LinearRegression

```python
# 변수 선택
X = df[['temp', 'humidity', 'windspeed', 'year', 'month', 'hour', 'dayofweek', 'holiday', 'workingday', 'weather']]
y = df['count']
```

```python
# 훈련세트/테스트세트 분할
from sklearn.model_selection import train_test_split # 훈련세트/테스트세트 분할
X_train, X_test, y_train, y_test = train_test_split(X, df['count'], test_size=0.25, random_state=1) # 25% 테스트세트
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
# 모델 생성
from sklearn.linear_model import LinearRegression # 선형회귀 모델
model = LinearRegression() # 모델 생성
model.fit(X_train, y_train) # 모델 훈련
```

```python
# 테스트데이터로 예측
pred = model.predict(X_test) # 테스트데이터로 예측

# 모델 성능 평가
from sklearn.metrics import mean_squared_error, root_mean_squared_error, mean_absolute_error, r2_score
mse = mean_squared_error(y_test, pred) # 평균제곱오차
rmse = np.sqrt(mse) # 평균제곱근오차
mae = mean_absolute_error(y_test, pred) # 평균절대오차
r2 = r2_score(y_test, pred) # 결정계수

train_r2 = model.score(X_train, y_train) # 훈련세트 결정계수
test_r2 = model.score(X_test, y_test) # 테스트세트 결정계수

print(f'''
    mse: {mse}
    rmse: {rmse}
    mae: {mae}
    r2: {r2}
    train_r2: {train_r2}
    test_r2: {test_r2}''')

# 시각화
plt.figure(figsize=(15,5))
plt.plot(y_test.values, label='실제값', marker='o')
plt.plot(pred, label='예측값', marker='x')
plt.legend()
```

## 원핫인코딩

```python
# 변수 선택
X = df[['temp', 'humidity', 'day','windspeed', 'year', 'month', 'hour', 'dayofweek', 'holiday', 'workingday', 'weather']]
y = df['count']

# 범주형변수 원핫인코딩
X_ohe = pd.get_dummies(X, columns=['year', 'month', 'hour', 'dayofweek', 'weather']) # 범주형변수 원핫인코딩 

display(X_ohe.head())
```

```python
# 훈련세트/테스트세트 분할
X_train, X_test, y_train, y_test = train_test_split(X_ohe, df['count'], test_size=0.25, random_state=42) # 25% 테스트세트
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
# 모델 생성
from sklearn.linear_model import LinearRegression # 선형회귀 모델
model = LinearRegression() # 모델 생성
model.fit(X_train, y_train) # 모델 훈련

 # 테스트데이터로 예측
pred = model.predict(X_test)

# 로그 역변환
# pred_original = np.expm1(pred) # 로그역변환
# y_test_original = np.expm1(y_test) # 로그역변환

# 모델 성능 평가
from sklearn.metrics import mean_squared_error, mean_absolute_error, root_mean_squared_error, r2_score
mse = mean_squared_error(y_test, pred) # 평균제곱오차
rmse = np.sqrt(mse) # 평균제곱근오차
mae = mean_absolute_error(y_test, pred) # 평균절대오차
r2 = r2_score(y_test, pred) # 결정계수

train_r2 = model.score(X_train, y_train) # 훈련세트 결정계수
test_r2 = model.score(X_test, y_test) # 테스트세트 결정계수

print(f'''
    MSE: {mse: .2f}
    RMSE: {rmse: .2f}
    MAE: {mae: .2f}
    R2: {r2: .2f}
    Train_R2: {train_r2: .2f}
    Test_R2: {test_r2: .2f}''')

model.score(X_train, y_train)
```

## 종속변수 로그변환

```python
# 변수 선택
X = df[['holiday', 'workingday', 'weather', 'temp', 'humidity', 'windspeed','year', 'month', 'hour', 'dayofweek']]
y = df['count']

# 범주형변수 원핫인코딩
X_ohe = pd.get_dummies(X, columns=['year', 'month', 'hour', 'dayofweek', 'weather']) # 범주형변수 원핫인코딩 

display(X_ohe.head())
```

```python
y_log = np.log1p(y)
sns.histplot(y_log)
```

```python
# 훈련세트/테스트세트 분할
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_ohe, y_log, random_state=42)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
 
 
# 모델 생성
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
 
 
# 테스트데이터로 예측
pred = model.predict(X_test)
 
# 로그 역변환
pred_original = np.expm1(pred)
y_test_original = np.expm1(y_test)
 
# 모델 성능 평가
from sklearn.metrics import mean_squared_error, mean_absolute_error, root_mean_squared_error, r2_score
mse = mean_squared_error(y_test_original, pred_original)
mae = mean_absolute_error(y_test_original, pred_original)
rmse = root_mean_squared_error(y_test_original, pred_original)
r2 = r2_score(y_test_original, pred_original)
 
print(f'''
mse : {mse}
mae : {mae}
rmse : {rmse}
r2 : {r2}
''')
 
model.score(X_train, y_train)
```

## cross validation

```python
from sklearn.model_selection import KFold, cross_val_score # KFold, cross_val_score 임포트

# KFold 객체 생성
kf = KFold(n_splits=5, shuffle=True, random_state=42) # 5개의 폴드로 나누기
print(cross_val_score(model, X, y, scoring='r2', cv=kf).mean()) # 교차검증 
print(cross_val_score(model, X_ohe, y, scoring='r2', cv=kf).mean()) # 교차검증 
print(cross_val_score(model, X_ohe, y_log, scoring='r2', cv=kf).mean()) # 교차검증 
```

# 다른 모델과 비교

```python
# 다른 모델과 비교
from sklearn.neighbors import KNeighborsRegressor # KNeighborsRegressor 임포트
model = KNeighborsRegressor() # 모델 생성

# KFold 객체 생성
kf = KFold(n_splits=5, shuffle=True, random_state=42) # 5개의 폴드로 나누기
print(cross_val_score(model, X, y, scoring='r2', cv=kf).mean()) # 교차검증 
print(cross_val_score(model, X_ohe, y, scoring='r2', cv=kf).mean()) # 교차검증 
print(cross_val_score(model, X_ohe, y_log, scoring='r2', cv=kf).mean()) # 교차검증 
```

```python
from sklearn.tree import DecisionTreeRegressor # DecisionTreeRegressor 임포트
model = DecisionTreeRegressor(max_depth=7) # 모델 생성

# KFold 객체 생성
kf = KFold(n_splits=5, shuffle=True, random_state=42) # 5개의 폴드로 나누기
print(cross_val_score(model, X, y, scoring='r2', cv=kf).mean()) # 교차검증 
print(cross_val_score(model, X_ohe, y, scoring='r2', cv=kf).mean()) # 교차검증 
print(cross_val_score(model, X_ohe, y_log, scoring='r2', cv=kf).mean()) # 교차검증 
```
